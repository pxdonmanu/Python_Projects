# Matplotlib Tutorial

##### Importing the libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

##### Read the data

```python
gas = pd.read_csv('gas_prices_data.csv')

gas
```

##### Line Chart

```python
# The gas price through the years in USA

# plt.plot(x, y)
plt.plot(gas.Year, gas.USA)
plt.show()
```

```python
# The gas price through the years in Mexico

plt.plot(gas.Year, gas.Mexico)
plt.show()
```

```python
# Comparation between two countries

plt.plot(gas.Year, gas.Canada)
plt.plot(gas.Year, gas.Germany)
plt.show()
```

```python
# Comparation between three countries

plt.plot(gas.Year, gas['South Korea'])
plt.plot(gas['Year'], gas['Japan'])
plt.plot(gas['Year'], gas.Italy)

plt.show()
```

##### Adding visual elements to the graph

```python
# Size Graph (firstly)
plt.figure(figsize=(12,6))

# Title graph
plt.title('Gas prices 1990-2008 (in USD)')

plt.plot(gas.Year, gas['Mexico'], 'g.-', label='Mexico') # 'g is green .- is for show the points through the years'
plt.plot(gas['Year'], gas['USA'], 'b.-', label='United States') # 'b is blue .- is for show the points through the years'
plt.plot(gas['Year'], gas.UK, 'r.-', label='United Kingdom') # 'r is red .- is for show the points through the years'
plt.plot(gas.Year, gas.Australia, 'y.-', label='Australia')

# Change the x text
plt.xticks(gas.Year)
### plt.xticks(gas.Year[::3]) To be every year but going in threes. 

# Label to x and y axes
plt.xlabel('Years')
plt.ylabel('Gas price')

plt.legend()

plt.show()
```

##### Line chart with all the countries

```python
plt.figure(figsize=(13,7))

plt.title('Gas Prices of all countries', fontdict={'fontweight':'bold', 'fontsize':20})

for country in gas:
  if country!='Year':
    plt.plot(gas.Year, gas[country], marker='.', label=country)

plt.xticks(gas.Year[::2])

plt.xlabel('Years')
plt.ylabel('Gas Price')

plt.legend()

# Save the graph
plt.savefig('graph_gas_countries.png', dpi=150)

plt.show()
```

##### Line chart graphing specific countries with a for loop

```python
plt.figure(figsize=(10,5))
plt.title('Selected Countries')

selected_countries=['Mexico', 'Australia', 'USA', 'Japan']

for countries in gas:
  if countries in selected_countries:
    plt.plot(gas.Year, gas[countries], marker='.', label=countries)

plt.xlabel('Year')
plt.ylabel('Gas Price')

plt.legend()

plt.show()
```
